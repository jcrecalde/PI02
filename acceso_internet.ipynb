Calculo la diferencia en un trimestre con respecto al anterior

```python
# Obtengo los valores de accesos por cada 100 hogares en un período anterior
accesos_anterior = df6.groupby(["Provincia"])['Accesos por cada 100 hogares'].shift(-1)
# Calculo el aumento del 2% en el período anterior
aumento = accesos_anterior * 1.02
# Asigno los valores calculados al trimestre siguiente
df6.loc[:, "Accesos por cada 100 hogares trimestre anterior + 2%"] = round(aumento,2)
```

```python
#Remplazo los valores nulos por los valores en la fila de la columna acceso por cada 100 hogares 
 
df6["Accesos por cada 100 hogares trimestre anterior + 2%"] = df6["Accesos por cada 100 hogares trimestre anterior + 2%"].fillna(df6["Accesos por cada 100 hogares"])
```

```python
#Ahora calculo la difencia 
 
df6["Diferencia"] = round(df6["Accesos por cada 100 hogares trimestre anterior + 2%"] - df6["Accesos por cada 100 hogares"], 2)
```

```python
#Luego calculo su porcentaje de firencia 

df6["Porcentaje de diferencia"] = round(df6["Diferencia"] / df6["Accesos por cada 100 hogares"] * 100, 2)
```
